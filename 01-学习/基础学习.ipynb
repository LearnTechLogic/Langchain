```python
from langchain_ollama.chat_models import ChatOllama
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory


def print_prompt(full_prompt):
    print("="*20,full_prompt.to_string(),"="*20)
    return full_prompt

model = ChatOllama(model="qwen3:4b")
prompt = PromptTemplate.from_template("你需要根据对话历史回应用户问题。对话历史：{chat_history}。用户当前输入：{input}， 请给出回应")

base_chain = prompt | print_prompt | model | StrOutputParser()

chat_history_store = {}
def get_history(session_id):
    if session_id not in chat_history_store:
        chat_history_store[session_id] = InMemoryChatMessageHistory()
    return chat_history_store[session_id]

conversion_chain = RunnableWithMessageHistory(
    base_chain,
    get_history,
    input_messages_key="input",
    history_messages_key="chat_history",
)
if __name__ == "__main__":
    session_cofig = {"configurable": {"session_id": "user_001"}}
    print(conversion_chain.invoke({"input": "小明有一只宠物"}, session_cofig))
    print(conversion_chain.invoke({"input": "小刚有二只宠物"}, session_cofig))
    print(conversion_chain.invoke({"input": "小王有三只宠物"}, session_cofig))
    print(conversion_chain.invoke({"input": "一共有几只宠物"}, session_cofig))
```

```python
from langchain_community.chat_message_histories import FileChatMessageHistory
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory, BaseMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import messages_from_dict, message_to_dict
from langchain_ollama.chat_models import ChatOllama
from typing import Sequence, List
import json
import os

model = ChatOllama(model="qwen3:4b")
prompt = PromptTemplate.from_template(
    """
    你是一个贴心的助手，需要根据对话历史回应用户的问题。
    对话历史：{chat_history}
    用户当前输入：{input}
    你的回应：
    """
)
base_chain = prompt | model | StrOutputParser()

def get_message_history(session_id: str) -> BaseChatMessageHistory:
    """
    根据会话ID获取对应的对话历史存储实例
    :param session_id: 会话唯一ID
    :return: FileChatMessageHistory 实例
    """
    # return FileChatMessageHistory(session_id=session_id, storage_path="./chat_history")
    # 创建目录
    os.makedirs("./chat_history", exist_ok=True)
    # 每个会话对应一个文件（session_id 作为文件名）
    file_path = f"./chat_history/{session_id}.json"
    return FileChatMessageHistory(file_path=file_path)

conversation_chain = RunnableWithMessageHistory(
    base_chain,
    get_session_history=get_message_history,
    input_messages_key="input",
    history_messages_key="chat_history",
)

if __name__ == "__main__":
    session_config = {"configurable": {"session_id": "user_002"}}
    response1 = conversation_chain.invoke({"input": "小明有一只宠物"}, config=session_config)
    print("第一轮",response1)
    response2 = conversation_chain.invoke({"input": "小刚有二只宠物"}, config=session_config)
    print("\n第二轮",response2)
    response3 = conversation_chain.invoke({"input": "小王有三只宠物"}, config=session_config)
    print("\n第三轮",response3)
    response4 = conversation_chain.invoke({"input": "一共有几只宠物"}, config=session_config)
    print("\n第四轮",response4)
```

```python
from langchain_core.documents import Document
from langchain_core.runnables import RunnablePassthrough
from langchain_ollama.chat_models import ChatOllama
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_ollama.embeddings import OllamaEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

model = ChatOllama(model="qwen3:4b")
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "以我提供的已知参考资料为主，简洁和专业的回答用户问题。参考资料:{context}"),
        ("user", "用户提问:{input}"),
    ]
)
vector_store = InMemoryVectorStore(embedding=OllamaEmbeddings(model="qwen3-embedding:4b"))

vector_store.add_texts(["Python是一个开源的计算机程序设计语言，用于创建可执行文件。", "Python是一种解释型语言，不需要编译就可以执行。", "Python是一种动态语言，可以在运行时修改变量的类型。", "Python是一种解释型语言，不需要编译就可以执行。", "Python是一种动态语言，可以在运行时修改变量的类型。", "Python是一种解释型语言，不需要编译就可以执行。", "Python是一种动态语言，可以在运行时修改变量的类型。", "Python是一种解释型"])

input_text = "Python是什么语言"

retriever = vector_store.as_retriever(search_kwargs={"k": 3})

def format_func(docs: list[Document]):
    if not docs:
        return "无相关参考资料"
    reference_text = "["
    for doc in docs:
        reference_text += doc.page_content + ";"
    reference_text += "]"
    return reference_text

# chain = retriever | prompt | model | StrOutputParser()
chain = (
    {"input": RunnablePassthrough(), "context": retriever | format_func} | prompt | model | StrOutputParser()
)
response = chain.invoke(input_text)
print(response)
```
